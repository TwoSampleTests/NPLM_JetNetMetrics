```python
import os
import subprocess
import sys
from collections import defaultdict
```

```python
results_dirs = ["results/jet_features/tests/10K",
                "results/jet_features/tests/10K_preprocessed",
                "results/jet_features/tests/20K",
                "results/jet_features/tests/20K_preprocessed",
                "results/jet_features/tests/50K",
                "results/jet_features/tests/50K_preprocessed",
                "results/particle_features_30/tests/10K",
                "results/particle_features_30/tests/10K_preprocessed",
                "results/particle_features_30/tests/20K",
                "results/particle_features_30/tests/20K_preprocessed",
                "results/particle_features_30/tests/50K",
                "results/particle_features_30/tests/50K_preprocessed"]
```

```python
preamble = r"""\documentclass[12pt]{article}
\pdfoutput=1

\usepackage{cite}
\usepackage{setspace}
\usepackage{xspace}
\usepackage{color}
\usepackage{soul}
\usepackage{amsmath, amsfonts, amssymb, amsthm, upgreek, amstext, bm, mathtools, mathrsfs, stmaryrd}
\usepackage[font=footnotesize,labelfont=bf]{caption}
\usepackage{slashed}
\usepackage{subfig}
\usepackage{multirow}
\usepackage{array} 
\usepackage{bbold}
\usepackage{physics}
\usepackage{booktabs}
\usepackage{fontawesome}
\usepackage[english]{babel}
\usepackage[utf8]{inputenc}
\usepackage[colorlinks=true,linkcolor=rossos,anchorcolor=black,citecolor=blue,filecolor=cyan,menucolor=red,runcolor=filecolor,urlcolor=blue,bookmarks=true,bookmarksnumbered=true]{hyperref}
\usepackage{cleveref}
\usepackage[dvipsnames]{xcolor}
\usepackage{doi}

\usepackage{graphicx}
\usepackage{pdflscape}

\setlength{\textheight}{9.in}
\setlength{\textwidth}{6.5in}
\setlength{\oddsidemargin}{0in}
\setlength{\topmargin}{-.6in}

\pagestyle{empty}

\begin{document}"""
```

```python
for dir in results_dirs:

    # Print info on the current directory
    print(f"Processing directory {dir}")

    # Read the latex table
    with open(os.path.join(dir,"results_table.tex"), 'r') as file:
        content = file.readlines()

    # Generate content of a compilable latex file
    latex_content = preamble+"\n"
    latex_content = latex_content + """\\begin{table*}[!h]
    \t\\centering
    \t\\small
    \t\\vspace{-4mm}\n\t"""
    latex_content += '\t'.join(content) + "\n"
    latex_content = latex_content+r"\end{table*}"+'\n'+"\end{document}"

    # Generate the latex file for compilation
    tmp_tex_file = os.path.join(dir, "results_table_tmp.tex")
    # Write the LaTeX content to the file
    with open(tmp_tex_file, 'w') as f:
        f.write(latex_content)

    # Compile the LaTeX file
    subprocess.run(['pdflatex', '-output-directory', dir, tmp_tex_file])

    # Delete auxiliary files
    aux_extensions = ['aux', 'log', 'out', 'toc']
    for ext in aux_extensions:
        aux_file = tmp_tex_file.replace('.tex', f'.{ext}')
        if os.path.exists(aux_file):
            os.remove(aux_file)
    if os.path.exists(tmp_tex_file.replace('.tex', '.pdf')):
        os.rename(tmp_tex_file.replace('.tex', '.pdf'), os.path.join(dir, "results_table.pdf"))
        os.remove(tmp_tex_file)
        
    print(f"Table for {dir} compiled")
        
print("All done!")
```
